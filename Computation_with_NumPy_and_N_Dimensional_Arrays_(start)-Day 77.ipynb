# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

**Challenge**: Let's multiply `a1` with `b1`. Looking at the wikipedia example above, work out the values for c12 and c33 on paper. Then use the [`.matmul()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) function or the `@` operator to check your work. 

```python
# Method 1: Using @ operator
result_at = a1 @ b1
print("Result using @ operator:\n", result_at)

# Method 2: Using np.matmul()
result_matmul = np.matmul(a1, b1)
print("Result using np.matmul():\n", result_matmul)

# Print specific values to verify c12 and c33
print("c12 =", result_at[0][1])  # should be 25
print("c33 =", result_at[3][2])  # should be 62
```

# Manipulating Images as ndarrays


```python
from PIL import Image

# Load the image
image = Image.open('/Users/puru/Desktop/landscape-of-green-rolling-hills-and-mountain-peaks.jpg')  # Replace with actual image file path
image_np = np.array(image)

# Show original image
plt.imshow(image_np)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Manipulation 1: Flip the image horizontally
flipped = np.fliplr(image_np)

# Manipulation 2: Darken the image
darker = (image_np * 0.5).astype(np.uint8)

# Show manipulated images
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].imshow(flipped)
axs[0].set_title("Horizontally Flipped")
axs[0].axis('off')

axs[1].imshow(darker)
axs[1].set_title("Darkened Image")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

**Challenge**: Convert the image to black and white. The values in our `img` range from 0 to 255. 
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
# Step 1: Normalize RGB values to range [0, 1]
img_srgb = image_np / 255.0

# Step 2: Define luminance weights for grayscale conversion
# These are the standard weights for sRGB luminance
grey_vals = np.array([0.2989, 0.5870, 0.1140])  # R, G, B

# Step 3: Perform grayscale conversion using dot product
grayscale_img = np.dot(img_srgb[...,:3], grey_vals)

# Step 4: Display grayscale image using matplotlib with colormap
plt.imshow(grayscale_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

#### Challenge Solutions

```python
# Flip grayscale vertically (upside down)
flipped_gray = np.flipud(grayscale_img)

# Display result
plt.imshow(flipped_gray, cmap='gray')
plt.title("Upside Down Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Rotate image 90° clockwise (k=-1)
rotated_color = np.rot90(image_np, k=-1)

# Display result
plt.imshow(rotated_color)
plt.title("Rotated Color Image (90° Clockwise)")
plt.axis('off')
plt.show()
```

```python
# Invert color image (solarize effect)
inverted_color = 255 - image_np

# Display result
plt.imshow(inverted_color)
plt.title("Inverted (Solarized) Color Image")
plt.axis('off')
plt.show()
```
